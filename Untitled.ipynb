```python
import numpy as np 
import pandas as pd
```

```python
customer = pd.read_csv("A_customer.csv")
film = pd.read_csv("A_film_revenue.csv")
geo = pd.read_csv("A_geography.csv")
inventory = pd.read_csv("A_inventory.csv")
staff = pd.read_csv("A_staff.csv")
time = pd.read_csv("A_time_base.csv")
```

```python
customer
```

```python
geo
```

```python
merge_1 = pd.merge(
    customer,
    geo,
    on=["city_id", "country_id" , "country", "city"],
    how="left",
)

merge_1
```

```python
merge_2 = pd.merge(
    merge_1,
    staff,
    on="store_id",
    how="left"
)

merge_2
```

```python
merge_2["merge_key"] = 1
inventory["merge_key"] = 1

merge_3 = pd.merge(
    merge_2,
    inventory,
    on="merge_key",
    how="left"
)

merge_3.drop("merge_key", axis=1, inplace=True)

merge_3
```

```python
merge_3["merge_key"] = 1
time["merge_key"] = 1

merge_4 = pd.merge(
    merge_3,
    time,
    on="merge_key",
    how="left"
)

merge_4.drop("merge_key", axis=1, inplace=True)

merge_4
```

```python
merge_4.to_csv("ABC.CSV")
```
